```python
#Transformation
#MinMaxScaler operation

from sklearn.preprocessing import MinMaxScaler

dummy=['Make','Model','Type','Trim']
df_2=pd.get_dummies(data[dummy],drop_first=True)
```

```python
#implementing the MinMaxScaler operation

scale=MinMaxScaler()

data["Mileage1"] = scale.fit_transform(data[["Mileage"]])
```

```python
data.head()
```

```python
data.drop(data[dummy],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop("Mileage",axis=1,inplace=True)
```

```python
data
```

```python
data.head()
```

```python
data_new= pd.concat([data,df_2], axis=1)
```

```python
data_new.head()
```

```python
data_new.shape
```

```python
#implementing Linear regression byu just taking one dependent and one independent variable
# By taking Mileage1 as independent variable and predicting the price values

x=data_new[["Mileage1"]]
y=data_new[["Price"]]
```

```python
#splitting the data into train and test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()

lr.fit(x_train, y_train)
```

```python
predict=lr.predict(x_test)
predict
```

```python
y_test
```

```python
#finding the error 

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,predict)
```
